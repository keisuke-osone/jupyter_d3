```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# csvファイルの読み込み
row_data = pd.read_csv('tokyo2016.csv', index_col=0)
```

```python
df = row_data.reset_index()
df['date'] = pd.to_datetime(df['date'], format='%Y/%m/%d')
df['power'] = df['power'].values.astype(int)
daily_data = df.groupby('date')['power'].sum().reset_index()
# daily_data = pd.DataFrame(daily_data['power'].values.astype(int), pd.date_range('2016-01-01', periods=len(daily_data['power']), freq='D'))
daily_data = daily_data.set_index(['date'])
daily_data.plot()
```

```python
import statsmodels.api as sm
res = sm.tsa.seasonal_decompose(daily_data.values, freq=7)
res.plot()
```
